```python
# If packages are missing, run this once in a cell:
# %pip install -U pandas matplotlib seaborn pyarrow

import os
from pathlib import Path
import pandas as pd

# Robust repo root finder so paths work from / or /src
def repo_root():
    p = Path.cwd()
    for _ in range(6):
        if (p/".git").exists() or (p/"requirements.txt").exists() or (p/"README.md").exists():
            return p
        p = p.parent
    return Path.cwd()

ROOT = repo_root()
DATA = ROOT / "data"
PARQUET = DATA / "events_history.parquet"
DB = ROOT / "events.db" 

print("ROOT:", ROOT)
print("PARQUET exists:", PARQUET.exists())
print("DB exists:", DB.exists())
```

```python
df = pd.read_parquet(PARQUET)  # requires pyarrow installed
print(f"{len(df):,} rows")
# Basic peek
df.head()
```
